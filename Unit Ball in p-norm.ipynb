```python
#plot how does a ball looks like in p-norm 
#1,2,5,0.5 norm are plotted
#given a radius of the norm ball r in type float
```

```python
#change radius here
r = 1
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def norm(x, p):
    #x should be a TWO ELEMENT TUPLE
    norm = (abs(abs(x[0])**p) + abs(abs(x[1])**p))**(1/p)
    return norm
```

```python
def normplt(points,p,r):
    plt.figure()
    normx = np.zeros((2,500))
    for i in range(0,len(points[0])):
        x = (points[0][i],points[1][i])
        inversenorm = 1 / norm(x,p)
        normx[0][i] = inversenorm * points[0][i] * r
        normx[1][i] = inversenorm * points[1][i] * r
    plt.plot(normx[0],normx[1])
    plt.gca().set_aspect('equal')
    fig_ = plt.gca()
    return fig_
```

```python
ps = [1, 2, 5, 0.5]
phi = np.linspace(0, 2 * np.pi, 500)
unit_circle = np.array([np.cos(phi), np.sin(phi)])
```

```python
#plot in 1 norm
fig_1 = normplt(unit_circle,1,r)
#plot in 2 norm
fig_2 = normplt(unit_circle,2,r)
#plot in 5 norm
fig_3 = normplt(unit_circle,5,r)
#plot in 0.5 norm
fig_4 = normplt(unit_circle,0.5,r)
```
