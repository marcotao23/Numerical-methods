```python
import numpy as np
import matplotlib.pyplot as plt

#call = lambda S: np.maximum(0, S - 40)
def call(S):
    return np.maximum(0,S-40)
#put = lambda S: np.maximum(0, 40 - S)
def put(S):
    return np.maximum(0,40-S)

n = np.asarray(range(0,80))

#call
plt.figure(1)
f1 = plt.plot(n,call(n),'-')
plt.title('call option payoff for strike price $40')
plt.xlabel('Price at Maturity')
plt.ylabel('payoff')

#put
plt.figure(2)
f2 = plt.plot(n,put(n),'-')
plt.title('put option payoff for strike price $40')
plt.xlabel('Price at Maturity')
plt.ylabel('payoff')
```
